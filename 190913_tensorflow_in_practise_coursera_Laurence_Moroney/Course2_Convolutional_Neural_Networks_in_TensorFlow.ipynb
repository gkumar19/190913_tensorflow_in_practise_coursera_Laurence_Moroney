Week1: Use image generator and concolutions with image argumentation, save accuracy

```
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import os
import tensorflow as tf

loc_base = "C:/Users/Gaurav/Desktop/Data_Science/Tensor_Flow/cats_and_dogs_filtered"
loc_train = loc_base + '/train'
loc_validation = loc_base + '/validation'

loc_train_cats = loc_train + '/cats'
loc_train_dogs = loc_train + '/dogs'

loc_validation_cats = loc_validation + '/cats'
loc_validation_dogs = loc_validation + '/dogs'

filename_train_cats = os.listdir(loc_train_cats)

img = []

for i in range(2):
    img.append(mpimg.imread(loc_train_cats + '/' + filename_train_cats[i]))
    plt.figure(i)
    plt.imshow(img[i])

model = tf.keras.models.Sequential([tf.keras.layers.Conv2D(16,(3,3), activation = 'relu',input_shape = (150,150,3)),
                                    tf.keras.layers.MaxPooling2D(2,2),
                                    tf.keras.layers.Conv2D(32,(3,3), activation = 'relu'),
                                    tf.keras.layers.MaxPooling2D(2,2),
                                    tf.keras.layers.Conv2D(64,(3,3), activation = 'relu'),
                                    tf.keras.layers.MaxPooling2D(2,2),
                                    tf.keras.layers.Flatten(),
                                    tf.keras.layers.Dense(512,activation = 'relu'),
                                    tf.keras.layers.Dense(1,activation = 'sigmoid')
                                    ])
model.summary()

model.compile(optimizer = tf.keras.optimizers.RMSprop(lr = 0.001),
              loss = tf.keras.losses.binary_crossentropy,
              metrics = ['acc'])

train_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale = 1.0/255.0)
validation_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale = 1.0/255.0)

train_generator = train_datagen.flow_from_directory(directory = loc_train,
                                                    batch_size = 20,
                                                    class_mode = 'binary',
                                                    target_size = (150,150))

validation_generator = train_datagen.flow_from_directory(directory = loc_validation,
                                                    batch_size = 20,
                                                    class_mode = 'binary',
                                                    target_size = (150,150))
model.load_weights('C:/Users/Gaurav/Desktop/Data_Science/Tensor_Flow/cats_and_dogs_filtered/weights.h5')

history = model.fit_generator(train_generator,
                              validation_data = validation_generator,
                              epochs = 1,
                              validation_steps = 50)

model.save_weights('C:/Users/Gaurav/Desktop/Data_Science/Tensor_Flow/cats_and_dogs_filtered/backup_weights.h5')

train_acc = history.history['acc']
train_loss = history.history['loss']

validation_acc = history.history['val_acc']
validation_loss = history.history['val_loss']

plt.figure()
plt.plot(train_loss)
plt.plot(validation_loss)
plt.figure()
plt.plot(train_acc)
plt.plot(validation_acc)

#Checking random images

image_check = tf.keras.preprocessing.image.load_img('C:/Users/Gaurav/Desktop/Data_Science/Tensor_Flow/cats_and_dogs_filtered/Check/dog1.jpg', target_size = (150,150))
image1 = tf.keras.preprocessing.image.img_to_array(image_check)
image2 = np.expand_dims(image1,axis = 0)
prediction = model.predict(image2)
if prediction == 1:
    print('its a dog')
else:
    print('its a cat')
```

Week3: Transfer Learning on inception model, load inception_V3 weights from: https://storage.googleapis.com/mledu-datasets/inception_v3_weights_tf_dim_ordering_tf_kernels_notop.h5

Also learn non-sequential keras modeling

```
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import os
import tensorflow as tf

pre_trained_model = tf.keras.applications.inception_v3.InceptionV3(input_shape = (150,150,3),
                                               include_top = False,
                                               weights = None)

pre_trained_model.load_weights('C:/Users/Gaurav/Desktop/Data_Science/Tensor_Flow/cats_and_dogs_filtered/inception_v3_weights.h5')


for layer in pre_trained_model.layers:
    layer.trainable = False

pre_trained_model.summary()

last_layer = pre_trained_model.get_layer('mixed7')
print('last layer shape is {}'.format(last_layer.output_shape))

last_output = last_layer.output

loc_base = "C:/Users/Gaurav/Desktop/Data_Science/Tensor_Flow/cats_and_dogs_filtered"
loc_train = loc_base + '/train'
loc_validation = loc_base + '/validation'

loc_train_cats = loc_train + '/cats'
loc_train_dogs = loc_train + '/dogs'

loc_validation_cats = loc_validation + '/cats'
loc_validation_dogs = loc_validation + '/dogs'

filename_train_cats = os.listdir(loc_train_cats)

img = []

for i in range(2):
    img.append(mpimg.imread(loc_train_cats + '/' + filename_train_cats[i]))
    plt.figure(i)
    plt.imshow(img[i])

x = tf.keras.layers.Flatten()(last_output)
x = tf.keras.layers.Dense(1024,activation = 'relu')(x)
x = tf.keras.layers.Dropout(0.2)(x)     #Is important to avoid overfitting in large model like this
x = tf.keras.layers.Dense(1, activation = 'sigmoid')(x)

model = tf.keras.Model(pre_trained_model.input, x)
model.summary()

model.compile(optimizer = tf.keras.optimizers.RMSprop(lr = 0.0001),
              loss = tf.keras.losses.binary_crossentropy,
              metrics = ['acc'])

train_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale = 1.0/255.0,
                                                                rotation_range = 40,
                                                                width_shift_range = 0.2,
                                                                height_shift_range = 0.2,
                                                                shear_range = 0.2,
                                                                zoom_range = 0.2,
                                                                horizontal_flip = True,
                                                                fill_mode = 'nearest')
validation_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale = 1.0/255.0)

train_generator = train_datagen.flow_from_directory(directory = loc_train,
                                                    batch_size = 20,
                                                    class_mode = 'binary',
                                                    target_size = (150,150))

validation_generator = train_datagen.flow_from_directory(directory = loc_validation,
                                                    batch_size = 20,
                                                    class_mode = 'binary',
                                                    target_size = (150,150))
#model.load_weights('C:/Users/Gaurav/Desktop/Data_Science/Tensor_Flow/cats_and_dogs_filtered/weights.h5')

history = model.fit_generator(train_generator,
                              validation_data = validation_generator,
                              epochs = 20,
                              validation_steps = 50)

model.save_weights('C:/Users/Gaurav/Desktop/Data_Science/Tensor_Flow/cats_and_dogs_filtered/backup_weights.h5')

train_acc = history.history['acc']
train_loss = history.history['loss']

validation_acc = history.history['val_acc']
validation_loss = history.history['val_loss']

plt.figure()
plt.plot(train_loss)
plt.plot(validation_loss)
plt.figure()
plt.plot(train_acc)
plt.plot(validation_acc)

#Checking random images

image_check = tf.keras.preprocessing.image.load_img('C:/Users/Gaurav/Desktop/Data_Science/Tensor_Flow/cats_and_dogs_filtered/Check/dog1.jpg', target_size = (150,150))
image1 = tf.keras.preprocessing.image.img_to_array(image_check)
image2 = np.expand_dims(image1,axis = 0)
prediction = model.predict(image2)
if prediction == 1:
    print('its a dog')
else:
    print('its a cat')
```

Week4: Multiclass classification, modelled using keras flexible API

```
# -*- coding: utf-8 -*-
"""
Created on Sun Aug  4 16:37:23 2019

@author: Gaurav
"""
# Import all important libraries
import tensorflow as tf
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import os

loc_training = "C:/Users/Gaurav/Desktop/Data_Science/Tensor_Flow/RPS/rps-train-set"
loc_validation = "C:/Users/Gaurav/Desktop/Data_Science/Tensor_Flow/RPS/rps-test-set"
loc_cross_validation = "C:/Users/Gaurav/Desktop/Data_Science/Tensor_Flow/RPS/cross_validation"

list_training = os.listdir(loc_training)
list_validation = os.listdir(loc_validation)

input_layer = tf.keras.layers.Input(shape = (150,150,3))
x = tf.keras.layers.Conv2D(64,(3,3),activation = 'relu')(input_layer)
x = tf.keras.layers.MaxPooling2D(2,2)(x)
x = tf.keras.layers.Conv2D(64,(3,3),activation = 'relu')(x)
x = tf.keras.layers.MaxPooling2D(2,2)(x)
x = tf.keras.layers.Conv2D(128,(3,3),activation = 'relu')(x)
x = tf.keras.layers.MaxPooling2D(2,2)(x)
x = tf.keras.layers.Conv2D(128,(3,3),activation = 'relu')(x)
x = tf.keras.layers.MaxPooling2D(2,2)(x)
x = tf.keras.layers.Flatten()(x)
x = tf.keras.layers.Dropout(0.2)(x)
x = tf.keras.layers.Dense(512, activation = 'relu')(x)
x = tf.keras.layers.Dense(3, activation = 'softmax')(x)

model = tf.keras.Model(inputs = input_layer ,outputs = x)

model.compile(optimizer = tf.keras.optimizers.RMSprop(lr = 0.001),
              loss = 'categorical_crossentropy',
              metrics = ['acc'])

model.summary()

training_generator = tf.keras.preprocessing.image.ImageDataGenerator(rescale = 1.0/255.0,
                                                                     rotation_range = 40,
                                                                     width_shift_range = 0.2,
                                                                     height_shift_range = 0.2,
                                                                     shear_range = 0.2,
                                                                     zoom_range = 0.2,
                                                                     horizontal_flip = True,
                                                                     fill_mode = 'nearest')
validation_generator = tf.keras.preprocessing.image.ImageDataGenerator(rescale = 1.0/255.0)

training_data = training_generator.flow_from_directory(loc_training,
                                                       target_size = (150,150),
                                                       class_mode = 'categorical',
                                                       batch_size = 50)
validation_data = validation_generator.flow_from_directory(loc_validation,
                                                       target_size = (150,150),
                                                       class_mode = 'categorical',
                                                       batch_size = 50)

history = model.fit_generator(training_data,
                    epochs = 20,
                    validation_data = validation_data,
                    validation_steps = 49)

plt.figure()
plt.plot(history.history['acc'],history.history['val_acc'] )
plt.figure()
plt.plot(history.history['loss'],history.history['val_loss'] )

image = tf.keras.preprocessing.image.load_img('C:/Users/Gaurav/Desktop/Data_Science/Tensor_Flow/RPS/cross_validation/rock2.png', target_size = (150,150))
image = tf.keras.preprocessing.image.img_to_array(image)
image = np.expand_dims(image, axis = 0)
prediction = model.predict(image)
```
